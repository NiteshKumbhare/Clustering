# Hierarchial Clusetring

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
```

```python
Crime=pd.read_csv("Downloads\\crime_data.csv")
Crime
```

```python
def minmax_func(i):
    x=(i-i.mean())/(i.max()-i.min())
    return(x)
```

```python
df_norm=minmax_func(Crime.iloc[:,1:])
df_norm
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_norm,method='complete'))
```

```python
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hc
```

```python
y_hc=hc.fit_predict(df_norm)
y_hc
```

```python
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters
```

```python
Clusters.value_counts()
```

```python
Crime['h_clusterid']=y_hc
Crime
```

```python
Crime.groupby('h_clusterid').mean()
```

# K-Means

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
Crime1=pd.read_csv("Downloads\\crime_data.csv")
Crime1
```

```python
from sklearn.preprocessing import StandardScaler
Scaler=StandardScaler()
Scaled_Crime1_df=Scaler.fit_transform(Crime1.iloc[:,1:])
Scaled_Crime1_df
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,random_state=0)
    kmeans.fit(Scaled_Crime1_df)
    wcss.append(kmeans.inertia_)
print(wcss)    
plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
            
```

```python
clusters_new=KMeans(4,random_state=0)
clusters_new.fit(Scaled_Crime1_df)
```

```python
clusters_new.labels_
```

```python
Crime1['clusterid_new']=clusters_new.labels_
Crime1
```

```python
Crime1.groupby('clusterid_new').agg(['mean']).reset_index()
```

```python
Crime1['clusterid_new'].value_counts()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Crime2=pd.read_csv("Downloads\\crime_data.csv")
Crime2
```

```python
print(Crime2.head())
```

```python
print(Crime2.info())
```

```python
Crime2.drop(['Unnamed: 0'],axis=1,inplace=True)
Crime2
```

```python
array=Crime2.values
array
```

```python
SS=StandardScaler()
x=SS.fit_transform(array)
x
```

```python
dbscan=DBSCAN(eps=1.2,min_samples=8)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
Crime2['Cluster']=pd.DataFrame(dbscan.labels_)
Crime2
```

```python
Crime2['Cluster'].value_counts()
```

```python
(len(Crime2[Crime2['Cluster']==-1])/len(Crime2))*100
```

```python
x=metrics.silhouette_score(x,dbscan.labels_)
x
```

```python
def dbscan(x,eps,min_samples):
    ss=StandardScaler()
    x=ss.fit_transform(x)
    db=DBSCAN(eps=eps,min_samples=min_samples)
    db.fit(x)
    y_pred=db.fit_predict(x)
    plt.scatter(x[:,0],x[:,1],c=y_pred,cmap='Paired')
    plt.title("DBSCAN")
```

```python
dbscan(array,1.0,8)
```
